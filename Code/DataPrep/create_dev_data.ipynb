```python
import pandas
from sklearn import model_selection


# Alterar "red" para "white" para rodar o mesmo para white wines
source_file   = '../../Data/Raw/winequality-red.csv'
save_dev_file =  '../../Data/Raw/winequality-red-dev.parquet'
save_op_file  =  '../../Data/Raw/winequality-red-op.parquet'
label_file = 'wine-red'

dev_percent = 0.8
min_quality = 5
```

# Carregar Dados Originais 

```python
data = pandas.read_csv(source_file, sep=';')
data['name'] = label_file + data.index.astype(str)
data.columns = [name.replace(' ','_') for name in data.columns]
print(data.shape)
data.head()
```

# Binarizar Fator de Qualidade 

```python
data['good_quality'] = data.quality > min_quality
data.groupby('good_quality').quality.count()
data.drop('quality',axis=1, inplace=True)
data.head()
```

# Separar para desenvolvimento

```python
Xdev, Xop, Ydev, Yop = model_selection.train_test_split(data.drop('good_quality', axis=1),
                                                        data.good_quality,
                                                        train_size = dev_percent,
                                                        stratify=data.good_quality.values) 
df_dev = pandas.concat((Xdev, Ydev) , axis=1)
df_op  = pandas.concat((Xop , Yop ) , axis=1).drop('good_quality',axis=1)

print('DEV', df_dev.shape)
print('OP', df_op.shape)
```

# Exportar os dados 

```python
df_dev.to_parquet(save_dev_file)
df_op.to_parquet(save_op_file)
```
